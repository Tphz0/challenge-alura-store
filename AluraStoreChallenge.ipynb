### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
from IPython.display import display
facturacion_data = []
for nombre, df in tiendas.items():
  facturacion_data.append({
      'tienda': nombre,
      'facturacion total': df['Precio'].sum()
  })
facturacion = pd.DataFrame(facturacion_data)
display(facturacion)
```

# 2. Ventas por categoría

```python
categoria_mas_popular_data = []
for nombre, df in tiendas.items():
  categoria_mas_popular_data.append({
      'tienda': nombre,
      'categoría más popular': df['Categoría del Producto'].mode()[0]
  })
  categoria_mas_popular = pd.DataFrame(categoria_mas_popular_data)
display(categoria_mas_popular)
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_data = []
for nombre, df in tiendas.items():
  promedio_calificacion_data.append({
      'tienda': nombre,
      'promedio de calificación': df['Calificación'].mean()
      })
promedio_calificacion = pd.DataFrame(promedio_calificacion_data)
display(promedio_calificacion)
```

# 4. Productos más y menos vendidos

```python
productos_vendidos_data = []
for nombre, df in tiendas.items():
  product_counts = df['Producto'].value_counts()
  productos_vendidos_data.append({
      'tienda': nombre,
      'producto más vendido': product_counts.idxmax(),
      'cantidad más vendida': product_counts.idxmin()
  })
  productos_mas_vendidos = pd.DataFrame(productos_vendidos_data)
display(productos_mas_vendidos)
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_data = []
for nombre, df in tiendas.items():
  costo_envio_promedio_data.append({
      'tienda': nombre,
      'costo envio promedio': df['Costo de envío'].mean()
  })
  costo_envio_promedio = pd.DataFrame(costo_envio_promedio_data)
display(costo_envio_promedio)
```

### 6. Resumen

```python
tiendas = {
  'Tienda 1': tienda,
  'Tienda 2': tienda2,
  'Tienda 3': tienda3,
  'Tienda 4': tienda4
}
resumen = facturacion.merge( categoria_mas_popular, on='tienda').merge(promedio_calificacion, on='tienda').merge(productos_mas_vendidos, on='tienda').merge(costo_envio_promedio, on='tienda')
display(resumen)
```
